```python
import nltk
import math
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer

from tqdm.notebook import tqdm, trange


#nltk.download()
def term_frq(term,document):
    counter = 0
    for word in document:
        if word ==term:
            counter+=1
    return counter/len(document)

def idf(terms,docs):
    terms_idf = {}
    number_of_docs = len(docs.keys())
    for t in terms:
        counter = 0
        for k in docs.keys():
            if t in docs[k]:
                counter+=1
        terms_idf[t] = math.log2(number_of_docs/counter)
    return terms_idf
        
stemmer = PorterStemmer()

#terms= []
#tf = {id:[[term,freq],...]}
#docs = {id:[term_token]}
#docs_frq = {id:[term_token:tf-idf]}

#term_idf = {term:idf_v,...}

docs = {}
terms = []
tf = {}
docs_frq = {}

c = tqdm(range(pd.DataFrame(df['Description'].unique(),columns = ['title']).shape[0]))
for index,row in pd.DataFrame(df['Description'].unique(),columns = ['title']).iterrows():
    if not (isinstance(row['title'],str)):
        continue
    sentance = row['title'].lower()
    tokens = word_tokenize(sentance)
    
    words = [word for word in tokens if word not in stopwords.words()]
    words = [stemmer.stem(word) for word in words]
    
    tf[index] = []
    for w in words:
        terms.append(w)
        tf[index].append([w,term_frq(w,words)])
    docs[index] = words
    c.update()
term_idf = idf(terms,docs)

for k in docs.keys():
    for k in tf.keys():
        for[term,freq] in tf[k]:
            docs_frq[k] = freq*term_idf[term]
            print(docs_frq[k])

```
